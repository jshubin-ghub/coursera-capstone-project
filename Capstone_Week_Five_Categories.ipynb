```python
def extract_categories(obj, nest, key1, key2):
    """recursively pull out nested ids, along with parent id, found in category json"""
    arr = []    
    def extract(obj, arr, nest, key1, key2, parent = None):
        new_parent = None    
        if isinstance(obj, dict):
            pair = []
            for k, v in obj.items():
                if k == key1:
                    pair.append(v)
                    print(v)
                    new_parent = v
                elif k == key2:
                    pair.append(v)
                    print(v)
                    if len(pair) == 2: #this is inelegant
                        pair.append(parent)
                        arr.append(pair)
                elif k == nest: #'categories' in this instance
                    for item in v:
                        extract(item, arr, nest, key1, key2, new_parent)
        elif isinstance(obj, list):
            for item in obj:
                extract(item, arr, nest, key1, key2, new_parent)
        return arr
    
    results = extract(obj, arr, nest, key1, key2)
    return results
```

```python
some_results = extract_categories(new_categories, 'categories', 'id', 'name')
```

```python
import pandas as pd
some_df = pd.DataFrame(some_results, columns = ['category_id','category_name', 'parent_id'])
some_df
```

```python
#print(some_df.loc[0]['category_id'])
# create a dictionary with category id keys and parent id values
category_dict = {}
for line in some_df.iterrows():
    category_dict[line[1][0]]=line[1][2]
category_dict
```

```python
# recursively find the highest level parent, creating a new dictionary of category id and top parent id

def find_parent(child):
    parent = category_dict[child]
    if parent==None:
        return child
    else:
        return(find_parent(parent))

top_category_dict = {}
for key in category_dict.keys():
    parent = find_parent(key)
    top_category_dict[key]=parent
```

```python
# add top parent id to the dataframe

for line in some_df.iterrows():
    some_df.loc[line[0],'top_parent_id']=top_category_dict[line[1][0]]
```
